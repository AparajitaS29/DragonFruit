```python
import numpy as np
import matplotlib.pyplot as plt
import time

#create a down sized matrix of size 1000 x 1000 pixels to show better visual representation on the notebook
x_max = 1000
y_max = 1000

#declare center coordinates for a 1000x1000 pixel image to generate a polygon shaped blob 
center = (500, 500)

# Define the number of vertices and the radius of the polygon shaped blob to be generated
num_vertices = np.random.randint(3, 10)  # Random number of vertices between 3 and 9
radius = np.random.randint(125, 300)   

xv, yv = np.meshgrid(np.arange(x_max), np.arange(y_max))

# Blob is represented as a binary array of 0s and 1s 
binary_array = np.zeros((x_max, y_max))
for i in range(num_vertices):
    angle = i * 2 * np.pi / num_vertices
    x = center[0] + radius * np.cos(angle)
    y = center[1] + radius * np.sin(angle)
    binary_array += ((xv - x) ** 2 + (yv - y) ** 2) <= radius ** 2


binary_array[binary_array > 0] = 1

#print(binary_array.astype(int))
```

```python
import numpy as np

# Declare a down sized size of an array
array_size = 1000

# fill the array entirely with 0s 
array1 = np.zeros((array_size, array_size))

# We are randomly filling the parasites in the array with 1s where minimum 25% is parasite with an upper limit of 40%
percentage_ones = np.random.randint(25,40)

# Calculate the number of parasites to fill
num_ones = int(array_size ** 2 * percentage_ones / 100)

# Generate a list of random coordinates
coordinates = np.random.randint(0, array_size, size=(num_ones, 2))

# The Dye is now randomly injected into parts of the array where there is a parasite along with some leak. 
array1[coordinates[:, 0], coordinates[:, 1]] = 2

#print(array.astype(int))
```

```python
#Add the values where there is a parasite and a dye is present which increases the sum count to 4 
def cancer_optimized(binary_array,array1):
    start_time = time.time()

    sum1 = np.add(binary_array,array1)
    ones_count = np.count_nonzero(binary_array==1) #Count of total number of parasite pixels in the environment
    twos_count = np.count_nonzero(sum1==3)
    res = (twos_count/ones_count)*100 
    if(res>10):
        print("The parasite is a cancerous parasite")
    else:
        print("The parasite is non cancerous")
    end_time = time.time()

    execution_time = end_time - start_time

    print("Execution time:", execution_time, "seconds")
```

```python
def cancer(binary_array, array1):
    start_time = time.time()

    binary_array = np.array(binary_array)
    array1 = np.array(array1)

    ones_count = np.count_nonzero(binary_array)
    twos_count = np.count_nonzero(np.logical_and(binary_array, array1))
    res = (twos_count/ones_count)*100 
    if(res > 10):
        print("The parasite is a cancerous parasite")
    else:
        print("The parasite is non cancerous")
    
    end_time = time.time()
    execution_time = end_time - start_time
    print("Execution time:", execution_time, "seconds")
```

Result is used to calculate the ratio of dye-parasite pixels to parasite pixels to detect cancer. If this count is greater than 10, then it has cancer
